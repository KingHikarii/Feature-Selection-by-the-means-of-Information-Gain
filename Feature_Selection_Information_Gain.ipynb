```python
import pandas as pd
nba_df = pd.read_csv('/content/Averaged.csv')
```

```python
float_list = ['float64']
float_columns = list(nba_df.select_dtypes(include=float_list).columns)
```

```python
float_columns
```

```python
nba_df = nba_df[float_columns]
```

```python
nba_df.head()
```

```python
from sklearn.model_selection import train_test_split 

#In order to avoid overfitting.

X_train, X_test, y_train, y_test = train_test_split(nba_df.drop(labels = ['WINP'], axis=1),
                                                    nba_df['WINP'],
                                                    test_size = 0.3,
                                                    random_state=0)
```

```python
X_train
```

```python
from sklearn.feature_selection import mutual_info_regression

mutual_info = mutual_info_regression(X_train.fillna(0), y_train) #Receiving the Information Gain of our Regression model
mutual_info
```

```python
mutual_info = pd.Series(mutual_info)
mutual_info.index = X_train.columns
mutual_info.sort_values(ascending=False) #The significnce of each of the independent variable
```

```python
mutual_info.sort_values(ascending = False).plot.bar(figsize=(15, 5)) #Graphical representation of each of the significance
```

```python
from sklearn.feature_selection import SelectPercentile
```

```python
#Selection of the Top 20 Percentiles.

top_column_selection = SelectPercentile(mutual_info_regression, percentile=20)
top_column_selection.fit(X_train.fillna(0), y_train)
X_train.columns[top_column_selection.get_support()]
```
